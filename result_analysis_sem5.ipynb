## Generate student ranklist on the basis of SGPA

```python
df_gpa_sorted = df.sort_values(by="SGPA",ascending=False)
df_gpa_sorted["Rank"] = df_gpa_sorted["SGPA"].rank(method="first",ascending=False)
df_gpa_sorted = df_gpa_sorted.reset_index(drop=True)
df_gpa_sorted
```

```python
ranks = []
for index,row in df_gpa_sorted.iterrows():
    local = []
    local.append(int(row["Rank"]))
    local.append(row["Name"])
    local.append(row["SGPA"])
    ranks.append(local)
ranks
```
